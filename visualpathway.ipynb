```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph to represent the visual pathway
G = nx.DiGraph()

# Add nodes representing different parts of the visual pathway
nodes = [
    "Eye (Retina)", 
    "Optic Nerve", 
    "Lateral Geniculate Nucleus (LGN)", 
    "Primary Visual Cortex (V1)"
]
G.add_nodes_from(nodes)

# Add edges representing normal signal flow
G.add_edges_from([
    ("Eye (Retina)", "Optic Nerve"),
    ("Optic Nerve", "Lateral Geniculate Nucleus (LGN)"),
    ("Lateral Geniculate Nucleus (LGN)", "Primary Visual Cortex (V1)")
])

# Add a bypass route to simulate direct stimulation of V1
G.add_edge("Eye (Retina)", "Primary Visual Cortex (V1)")

# Define node colors to represent damage and restoration
node_colors = ["red" if node == "Eye (Retina)" else "green" for node in nodes]
edge_colors = ["red" if "Retina" in edge[0] and "Optic Nerve" in edge[1] else "blue" for edge in G.edges]

# Visualize the network
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color=node_colors, edge_color=edge_colors, node_size=3000, font_size=10)
plt.title("Visual Pathway Simulation")
plt.show()
```
